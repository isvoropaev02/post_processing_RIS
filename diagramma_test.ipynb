```python
import numpy as np
import pandas as pd
import cmath

%matplotlib widget
import matplotlib.pyplot as plt
```

```python
diag = pd.read_csv('C:\\Users\\isvoropaev\\Desktop\\python_codes\\post_processing_RIS\\diagramma.txt',
                                  sep=' ', header=None, skiprows=2, index_col=False, skipinitialspace=True,
                                  names=['angle', '1', 'abs', '2', '3', '4', '5', '6'])
```

```python
diag = diag.drop(['1', '2', '3', '4', '5', '6'], axis=1)
```

```python
fig4 = plt.figure(figsize=(8, 5))

ax_1 = fig4.add_subplot(111)

ax_1.plot(diag['angle'], diag['abs']/(np.max(diag['abs'])), color='r')
ax_1.set_xlabel("Angle, deg")
ax_1.set_ylabel(r"Normalized Power flow")
ax_1.grid()


fig4.suptitle(r'Scattering pattern')
plt.show()
```

```python
mes_amp_db = np.array([-35, -31,-30, -30, -37, -62, -80, -85, -86, -81])
mes_angles = np.array([15, 17, 19, 21, 23, 26, 35, 45, 55, 60])

mes_amp = 10**(mes_amp_db/20)
mes_amp /= np.max(mes_amp)
```

```python
fig5 = plt.figure(figsize=(8, 5))

ax_1 = fig5.add_subplot(111)

ax_1.plot(diag['angle'], diag['abs']/(np.max(diag['abs'])), color='r', label='')
ax_1.scatter(mes_angles, mes_amp, label=' ')
ax_1.set_xlabel(", .")
ax_1.set_ylabel(r"  ")
ax_1.set_xlim([0, 70])
ax_1.grid()


#fig5.suptitle(r'Scattering pattern')
plt.show()
```

```python
fig5 = plt.figure(figsize=(8, 5))

ax_1 = fig5.add_subplot(111)

ax_1.plot(diag['angle'], 20*np.log10(diag['abs']/(np.max(diag['abs']))), color='r', label='')
ax_1.scatter(mes_angles, 20*np.log10(mes_amp), label=' ')
ax_1.set_xlabel(r"$\psi$, .")
ax_1.set_ylabel(r"   $F(\psi)$")
ax_1.set_xlim([0, 70])
ax_1.set_ylim(bottom=-75)
ax_1.legend()
ax_1.grid()


#fig5.suptitle(r'Scattering pattern')
plt.show()
```
